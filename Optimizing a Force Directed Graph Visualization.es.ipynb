```javascript
// https://stackoverflow.com/a/12588826/14390472
function createNDimArray(fn, dimensions) {
  if (dimensions.length > 0) {
    var dim = dimensions[0];
    var rest = dimensions.slice(1);
    var newArray = new Array();
    for (var i = 0; i < dim; i++) {
      newArray[i] = createNDimArray(fn, rest);
    }
    return newArray;
  } else {
    return fn();
  }
}

var vertices = createNDimArray(() => Math.random(), [10, 10, 10, 4]);
vertices
```

```javascript
Array.prototype.recursiveMap = function(fn, indices, level=0){
  if(!indices){
    indices = Array(this.length-1).fill(0);
  }
  
  return this.map(function(item, i, array){
    if(item instanceof Array){
      item = item.map((item, j, arrayInner) => fn(item, indices, arrayInner, level + 1));
    }else{
      item = fn(item, indices, array, level + 1);
    }
    
    return item;
  })
}

Array.prototype.index = function(...indices){
  var item = this;
  for(var i=0; i<indices.length; i++){
    item = item[indices[i]];
  }
  
  return item;
}

vertices = vertices.recursiveMap(function(item, indices){
  return item;
})
```

```javascript
var total = [];
var all = vertices.map((X, ix) =>
  X.map((Y, iy) =>
    Y.map((Z, iz) => {
	  return {indices: [ix, iy, iz], id: Z[0], x: Z[1], y: Z[2], z: Z[3]}
})))

all = all.flat().flat().flat()

var copy = (arr) => JSON.parse(JSON.stringify(arr));
var all1 = copy(all);
var all2 = copy(all);
var all3 = copy(all);
```

```javascript
function *runningAvg() {
  var lastSum = 0.0;
  var n = 0;
  var runAvg = 0.0;
  
  while(true){
    n++;
    var input = yield runAvg;;
    lastSum += input;
    runAvg = (1.0/n) * lastSum;
  }
}

var avg = runningAvg();
console.log(avg.next(0.0))
var results = [];
for(var i=0; i<10; i++){
  results.push(avg.next(i).value)
}

results;
```
