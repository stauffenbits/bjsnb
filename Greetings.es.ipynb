Jupyter Notebooks famously run in the browser. This is great: The sys-admin doesn't have to worry about what you're running, as long as it can run an HTML5-capable browser.

But if you write [Python 3](https://python.org), [Node.js](https://nodejs.org), [Julia](https://julialang.org), or even [box-lang](https://github.com/box-lang), you might have wondered: 

> Why isn't there a kernel for writing vanilla Javascript?

Well, off the top of my head, there are good reasons for that. Jupyter runs interpreted languages, and acts as a runtime for relaying commands and results between kernels running on the server, and presentation logic running in your browser. Or noone's gotten around to it yet. 

But don't fret! Thanks to the ingenuity of ***fine bitstrings***, you can now run small javascript experiments in the browser, document them, and show off your insights in a browser-javascript jupyter-like notebook.
